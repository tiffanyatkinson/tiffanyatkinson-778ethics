```python
class Time:
    @classmethod
    def fromcomponents(cls, hours, minutes, seconds):
        obj = cls()
        obj.hours = hours
        obj.minutes = minutes
        obj.seconds = seconds
        return obj

    def toseconds(self):
        return self.hours * 3600 + self.minutes * 60 + self.seconds

def subtracttime(timeone, timetwo):
    secondsone = timeone.toseconds()
    secondstwo = timetwo.toseconds()
    

    return abs(secondsone - secondstwo)

timea = Time.fromcomponents(2, 45, 36)  
timeb = Time.fromcomponents(12, 20, 59) 

interval = subtracttime(timea, timeb)
print(f"The interval is {interval} seconds.")
```

```python
class Time:
    @classmethod
    def fromcomponents(cls, hours, minutes, seconds):
        obj = cls()
        obj.hours = hours
        obj.minutes = minutes
        obj.seconds = seconds
        return obj

    def toseconds(self):
        return self.hours * 3600 + self.minutes * 60 + self.seconds

def isafter(time1, time2):
    return time1.toseconds() > time2.toseconds()

timea = Time.fromcomponents(2, 45, 36)  
timeb = Time.fromcomponents(12, 20, 59)  

result = isafter(timea, timeb)
print(f"Is timea after timeb? {result}")
```

```python
class Date:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    def __str__(self):
        return f"{self.year:04d}-{self.month:02d}-{self.day:02d}"

    def totuple(self):
        return (self.year, self.month, self.day)

    def isafter(self, other):
        return self.totuple() > other.to_tuple()

dateone = Date(1933, 6, 22)
print(dateone)  

datetwo = Date(1933, 9, 17)

result = datetwo.isafter(date1)
print(f"Is {dateone} after {datetwo}? {result}")
```
